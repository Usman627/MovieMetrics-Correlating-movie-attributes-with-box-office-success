```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'C:\Users\Hp\Downloads\movies.csv')
```

```python
for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_pct*100)))
```

```python
df.dtypes
```

```python
df['gross'] = df['gross'].astype('int64')
```

```python
df.drop_duplicates()
```

```python
df.sort_values(by=['year'], inplace=False, ascending=True)
```

```python
sns.regplot(x='gross', y='budget', data=df)
```

```python
df.corr(method ='pearson')
```

```python
correlation_matrix = df.corr(method ='kendall')
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
plt.show()
```

```python
sns.regplot(x='gross', y='votes', data=df)

plt.title("Regression plot for gross vs votes")

plt.xlabel("Gross")

plt.ylabel("Votes")
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.show()
```

```python

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)


sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)

```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) < 0.1]

print(strong_pairs)
```

```python


companygrosssum = df.groupby('year')[["gross"]].sum()
companygrosssumsorted = companygrosssum.sort_values('gross', ascending=False)[:10]
companygrosssumsorted = companygrosssumsorted['gross'].astype('int64')

companygrosssumsorted
```

```python
#multiple groupby

companyandyear = df.groupby(['company', 'year'])[["gross"]].sum()

candysorted = companyandyear.sort_values('gross', ascending=False)[:10]
candysorted = candysorted['gross'].astype('int64')

candysorted
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('gross vs budget')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```
